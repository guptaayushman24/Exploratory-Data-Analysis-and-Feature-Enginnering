```python
def outlier (df,variable) :
    q3 = df[variable].quantile(0.75)
    q1 = df[variable].quantile(0.25)
    iqr = q3-q1
    lower_fence = q1-1.5*iqr
    higherfence = q3+1.5*iqr
    return lower_fence,higherfence
```

```python
aggregate_rating = outlier(data,'Aggregate rating')
```

```python
# Function for plotting the pie for each top 11 city and their price range
def rename(list) :
  for i in list :
      if (i==1) :
            a=list.index(i)
            list[a] = 'Inexpensive'
      elif (i==2) :
          b=list.index(i)
          list[b] = 'Average'
        
      elif (i==3) :
          c=list.index(i)
          list[c] = 'Expensive'
      elif (i==4) :
          d=list.index(i)
          list[d] = 'Very Expensive'
  return list

def lst_counts(df,City) :
   lst_counts = df[df['City'] == City]['Price range'].value_counts().to_list()
   return (lst_counts)


def range_list (df,City) :
  lst_range = df[df['City'] == City]['Price range'].value_counts().index.to_list()
  return (rename(lst_range))

def plt_pie(df,City) :
  lst1=lst_counts(df,City)
  lst2=range_list(df,City)
  fig, ax = plt.subplots()
  ax.pie(lst1, labels=lst2)
  total = sum(lst1)
  plt.title('Pie Chart of {}'.format(City))
  plt.pie(lst1, labels=lst2 , autopct=lambda p: f'{p:.1f}% ({int(p * total / 100)})')
```
