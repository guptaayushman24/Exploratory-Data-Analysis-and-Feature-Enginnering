```python
# Storing all the name of the columns in the list
lst_columns=[]
for i in df.columns :
    lst_columns.append(i)
```

```python
lst_num=[]
lst_cat=[]
for i in lst_columns :
    if (df[i].dtypes==object) :
        lst_cat.append(i)
    else :
        lst_num.append(i)
```
