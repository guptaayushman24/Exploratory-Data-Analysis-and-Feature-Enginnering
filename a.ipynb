```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import LabelEncoder
import pandas as pd

class CustomEncoder(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        # Fit a label encoder for each categorical column (excluding 'CarName')
        self.encoders = {}
        for col in X.columns:
            if col != 'CarName' and X[col].dtype == 'object':
                encoder = LabelEncoder()
                encoder.fit(X[col])
                self.encoders[col] = encoder
        return self

    def transform(self, X):
        X_encoded = X.copy()
        # Apply label encoding to categorical columns
        for col, encoder in self.encoders.items():
            if col != 'CarName' and col in X_encoded.columns:
                X_encoded[col] = encoder.transform(X_encoded[col])
        # Drop 'CarName' column
        X_encoded.drop(['CarName'], axis=1, inplace=True)
        return X_encoded

# Usage:
# Assuming df is your DataFrame
# encoder = CustomEncoder()
# df = encoder.fit_transform(df)
```

```python
from sklearn.pipeline import Pipeline
pipe = Pipeline([
    ('encoder',CustomEncoder())
])
df = pipe.fit_transform(df)
```
