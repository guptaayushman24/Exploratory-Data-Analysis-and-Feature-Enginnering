## Feature Selction -> Dropping constant feature

In this step will be removing the feature which have constant features which are actually for solving the problem statement

```python
import pandas as pd
data = pd.DataFrame({"A":[1,2,4,1,2,4],
                    "B":[4,5,6,7,8,9],
                    "C":[0,0,0,0,0,0],
                    "D":[1,1,1,1,1,1]})
```

```python
data.head()
```

## Variance Threshold

Feature selector that removes all low-variance features.

This feature selection algorithm looks only at the features (X), not the desired outputs (y), and can thus be used for unsupervised learning.

```python
## It will zero variance features
from sklearn.feature_selection import VarianceThreshold
var_thres = VarianceThreshold(threshold=0)
var_thres.fit(data)
```

```python
var_thres.get_support()
```

```python
data_list = data.columns[var_thres.get_support()]
```

```python
data_list
```

```python
constant_columns=[]
for i in data.columns :
    if (i!='A' and i!='B') :
        constant_columns.append(i)
```

```python
len(constant_columns)
```

```python
constant_columns
```

```python
data.drop(constant_columns,axis=1,inplace=True)
```
