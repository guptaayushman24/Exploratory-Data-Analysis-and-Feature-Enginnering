# Handling and Visualizing the outliers in the dataset

Visualizing the ouliers using the box plot

```python
for i in lst_num :
    sns.boxplot(data=data,x=data[i])
    plt.show()
```

```python
lst_num
```

```python
# Creating the function for the outliers
def outliers (df,variable) :
    q3 = df[variable].quantile(0.75)
    q1 = df[variable].quantile(0.25)
    iqr = q3-q1
    lower_range = q1-1.5*iqr
    higher_range = q3+1.5*iqr
    return lower_range,higher_range
```

```python
# Checking the outliers in the numerical columns
for i in lst_num :
    l_range,h_range = outliers(data,i)
    print("Outliers in the {} column is from {} to {} ".format(i,l_range,h_range))
    
```

```python
for i in lst_num :
    feature_name = i

    l_range, h_range = outliers(data, feature_name)

    outliers_df = data[(data[feature_name] < l_range) | (data[feature_name] > h_range)]
    total_outliers = len(outliers_df)

    print("Total number of outliers in the {} feature: {}".format(i, total_outliers))
```

Applying one-hot encoding technique in the categorical column

```python
# Applying the one-hot encoding on the has_job_experience replacing Y with 1 and N with 0
data ['has_job_experience'] = data['has_job_experience'].map({'Y':1,'N':0})
```

```python
data.head()
```

```python
# Applying the one-hot encoding on the requires_job_training replacing Y with 1 and N with 0
data ['requires_job_training'] = data['requires_job_training'].map({'Y':1,'N':0})
```

```python
# Applying the one-hot encoding on the requires_job_training replacing Y with 1 and N with 0
data ['full_time_position'] = data['full_time_position'].map({'Y':1,'N':0})
```

```python
# Applying the one-hot encoding on the continent replacing Asia with 0,Africa  with 1,North America with 2, Europe with 3 , South America with 4 and Oceania with 5
data['continent'] = data['continent'].map({'Asia':0,'Africa':1,'North America':2,'Europe':3,'South America':4,'Oceania':5})
```

```python
data.head()
```

```python
# Applying the one-hot encoding on the region_of_employment replacing West with 0,Northeast with 1,South with 2,Midwest with 3 and Island with 4
data['region_of_employment'] = data['region_of_employment'].map({'West':0,'Northeast':1,'South':2,'Midwest':3,'Island':4})
```

```python
# Applying the one-hot encoding on the unit_of_wage replacing Hour with 0,Year with 1,Week  with 2,Month with 3 and Island with 4
data['unit_of_wage'] = data['unit_of_wage'].map({'Hour':0,'Year':1,'Week':2,'Month':3})
```
